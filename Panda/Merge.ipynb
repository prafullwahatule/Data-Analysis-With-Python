# Merge

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key':['x','y'],
                  'A':['A1','A2'],
                  'B':['B1','B2']})
df1.set_index('key', inplace = True)
df1
```

```python
df2 = pd.DataFrame({'key':['x','z'],
                  'A':['A1','A2'],
                  'C':['C1','C2']})
df2.set_index('key', inplace = True)
df2
```

#### When to use on=

```python
df1.merge(df2, on='A', how='left')
```

#### When to use left_index=True

```python
df1.merge(df2, left_index=True, right_on='A', how='left')
```

#### When both sides use index

```python
df1.merge(df2, left_index=True, right_index=True, how='left')
```
