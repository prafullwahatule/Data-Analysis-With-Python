```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generate 50 rows of dummy sales data
np.random.seed(42)

products = ["Apple", "Banana", "Orange", "Grapes", "Mango", "Pineapple"]
regions = ["East", "West", "North", "South"]

dates = [datetime.today() - timedelta(days=i) for i in range(50)]

df = pd.DataFrame({
    "Date": [d.strftime("%Y-%m-%d") for d in dates],
    "Product": np.random.choice(products, 50),
    "Quantity": np.random.randint(1, 50, 50),
    "Price": np.random.randint(50, 500, 50)
})

df["Total"] = df["Quantity"] * df["Price"]
df["Region"] = np.random.choice(regions, 50)

file_path = "C:/Users/Prafull Wahatule/Desktop/Python_DA/Data/sales_data_50rows.csv"
df.to_csv(file_path, index=False)

file_path
```

#### Q1. What percentage of the order total does each product represent?

```python
totalamount = df.groupby("Product")["Total"].sum().rename("Total Amount").reset_index()
```

```python
totalamount
```

```python
df_merge = df.merge(totalamount)
df_merge
```

```python
df_merge["% of Order By Product"] = df_merge["Total"] / df_merge["Total Amount"]
```
