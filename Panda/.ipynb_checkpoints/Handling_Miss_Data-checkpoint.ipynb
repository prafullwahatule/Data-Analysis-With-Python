### Get Count of Missing Data

```python
df.isnull().sum()  # it give you count of all null value in data 
```

### df.fillna(method='ffill', inplace=True)Filling Missing Data

#### 1.Forword fill
#### 2.Backword fill
#### 3.Forword Fill with a specific value
#### 4.Froword Fill only specific column

```python
df.ffill(inplace=True)   # F mean forword fill  = ffill
```

```python
df.bfill(inplace=True)  # B mean Backword fill = bfill
```

```python
df.fillna(0, inplace=True)  # Fill with a specific value
```

```python
df["Mg"] = df["Mg"].ffill() # Fill only specific column
```
