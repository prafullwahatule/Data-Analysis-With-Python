# Group By

```python
import pandas as pd
```

```python
winedata = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\wine.csv")
```

```python
print(winedata)
```

### How to Create Group Object

```python
dfwine = winedata.groupby(['Wine'])
dfwine
```

### Frequently Used Group by Operations

```python
dfwine.first()
```

```python
dfwine['Ash'].max()
```

```python
dfwine['Ash'].min()
```

```python
dfwine['Ash'].mean()
```

### Iterating Groups

```python
dfwine.groups
```

```python
for wine,wine_df in dfwine:
    print(wine)
    print(wine_df)
```

### How to Get Information About a Specific Group

```python
dfwine.get_group((1,))
```

### Getting Detailed View of the Grops

```python
dfwine.describe()
```

```python
dir(dfwine)
```

### Check the Length of Group

```python

len(dfwine.get_group((1,)))
```

```python
len(dfwine)
```
