# DataFrame Reshape

```python
import pandas as pd
```

```python
dictdf = {"EmpName" : ["Jacob","Marcos","Dane"],
     "EmpID" : [111,222,333],
     "EmpTitle" : ["Banker","Team Manager","Analyst"]}
df = pd.DataFrame(dictdf)
```

```python
df
```

```python
df_melted = pd.melt(df, 
                    id_vars=["EmpID"], 
                    value_vars=["EmpName", "EmpTitle"])

df_melted
```

```python
df_melted1 = pd.melt(df, 
                    id_vars=["EmpID","EmpName"], 
                    value_vars=["EmpTitle"])

df_melted1
```

```python
df_melted2 = pd.melt(df, 
                    id_vars=["EmpName"], 
                    value_vars=["EmpTitle"])

df_melted2
```

### Stack

```python
df_stack = df.stack()
df_stack
```

### Unstack

```python
df_unstack = df.unstack()
df_unstack
```
