# Window Functions

```python
import pandas as pd
import numpy as np
```

```python
price = np.random.randn(10,1)
days = pd.date_range('1/1/2025', periods = 10)
```

```python
df = pd.DataFrame(price, index = days, columns = ['StockPrice'])
df
```

```python
df["Prev_Price"] = df["StockPrice"].shift(1)
df
```

```python
df["Daily_Return"] = df["StockPrice"] / df["Prev_Price"]
df
```

```python
df["Expand_Mean"] = df["Daily_Return"].expanding().mean()
df
```

```python
df["Roll_Mean_3"] = df["Daily_Return"].rolling(3).mean()
df
```
