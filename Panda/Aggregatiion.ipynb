```python
df_gb = fruitsdata.groupby(['Type'])
df_gb
```

### What is Aggregation. How to do Aggregation in Pandas.

```python
df_gb.size()
```

```python
df_gb['Price'].mean()
```

### How Apply multiple Aggregation functions at once

```python
df_gb.agg([np.sum, np.mean, np.std])
```

### Renaming the Column names for aggregate functions

```python
df_gb.agg([np.sum, np.mean, np.std]).rename(column={
    "sum":"Count",
    "mean":"Avg",
    "std":"STD",
})
```
