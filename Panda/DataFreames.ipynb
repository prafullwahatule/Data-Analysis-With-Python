### Create DataFrames using List

```python
fruitlist = ["Apple", "Orange", "Grapes", "Banana"]  # List
dflist = pd.DataFrame(fruitlist) # Convert list into DataFrame
print(dflist) # Printing DataFram
type(dflist) # Check Type for Satisfaction
```

### Creating Multiple column DataFrames

```python
## If you want to add one another coloumn using list you need to create one more list
fruitlist1 = ["Apple", "Orange", "Grapes", "Banana"]  # List1
price1 = [100, 240, 150, 130]  # list2

dflist1 = pd.DataFrame(list(zip(fruitlist1,price1))) # Convert list into DataFram with two Column
print(dflist1) # print DataFrame
```

```python
## If you want to add one another coloumn using list you need to create one more list
## With custom column name
fruitlist1 = ["Apple", "Orange", "Grapes", "Banana"]  # List1
price1 = [100, 240, 150, 130]  # list2

dflist1 = pd.DataFrame(list(zip(fruitlist1,price1)), columns = ['Fruits','Price']) # Convert list into DataFram with two Column
print(dflist1) # print DataFrame
```
