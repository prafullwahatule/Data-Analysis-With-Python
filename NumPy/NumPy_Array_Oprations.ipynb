## Arthmetic Operations on NumPy Arrray

```python
arr1 = np.random.randint(1,10, size = (3,3))
```

```python
print(arr1)
```

```python
arr2 = np.random.randint(10,20, size = (3,3))
```

```python
print(arr2)
```

```python
np.add(arr1,arr2)
```

```python
np.subtract(arr1,arr2)
```

```python
np.multiply(arr1,arr2)
```

```python
np.divide(arr1,arr2)
```

```python
np.dot(arr1,arr2)
```
