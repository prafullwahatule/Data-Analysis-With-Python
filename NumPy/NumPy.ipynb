## What is Numpy

Definition

NumPy is a library that helps you work with arrays (like lists but faster) and perform mathematical operations quickly.

# Key Features of NumPy
1. Provides N-dimensional array object (ndarray)

A list in Python is slow.

A NumPy array is faster, uses less memory, and supports mathematical operations directly.

2. Fast mathematical calculations

Examples:

addition

multiplication

statistics (mean, median, std)

matrix operations

3. Used in Data Science & Machine Learning

Libraries like Pandas, Scikit-learn, and TensorFlow internally use NumPy.
