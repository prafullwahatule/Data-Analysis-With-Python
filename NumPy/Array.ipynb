## Array

An array is a data structure that stores multiple values in a single variable, all of the same type, arranged in a continuous (sequential) memory location.

In simple words:

üëâ Array = Collection of similar items stored together.

‚≠ê Features of an Array

- Stores multiple values under one name

- Values are stored in continuous memory

- Items are accessed using index numbers

- Index starts from 0

- Fast to access and modify

üìå Example of Array (Python ‚Äì List used as array)
fruits = ["apple", "banana", "orange"]
print(fruits[0])   # apple

üìå Example of NumPy Array
import numpy as np

arr = np.array([10, 20, 30, 40])
print(arr[1])  # 20

üß† Real-world Example of Arrays

- A list of students‚Äô marks

- A collection of numbers

- Pixel values of an image

- Rows and columns of Excel data

üìù Types of Arrays
1. One-dimensional Array

Example: [1, 2, 3, 4]

2. Two-dimensional Array (Matrix)

Example:

[ [1, 2, 3],
  [4, 5, 6] ]

3. Multi-dimensional Arrays

Used in ML, images, deep learning.

Example:

[ [ [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9] ] ]
