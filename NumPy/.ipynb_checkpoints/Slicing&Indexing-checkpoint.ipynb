# Slicing And Indexing

## 1D Array

```python
import numpy as np
```

```python
array1d = np.arange(10)
print(array1d)
```

### Extract Even Numbers From one Dim array

```python
even = array1d[:10:2]
print(even)
```

### Indexing

```python
array1d[5]
```

```python
array1d[::-1]
```

```python
array1d[-1]
```

```python
array1d[-3]
```

```python
array1d[-9]
```

## 2D Array

```python
array2d = np.random.randint(30,50, size = (2,2))
print(array2d)
```

```python
array2d.ndim
```

```python
array2d[1:]
```

```python
array2d[0:2,0:1]
```

```python
array2d[0],[0]
```

```python
array2d[-1],[-1]
```

## 3D Array

```python
array3d = np.random.randint(20,90, size = (1,3,4))
print(array3d)
```

```python
array3d.ndim
```

```python
array3d[1:]
```

```python
array3d[0:2,0:1]
```
