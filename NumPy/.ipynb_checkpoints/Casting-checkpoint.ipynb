# Casting 

# Upcasting & Downcasting

```python
import numpy as np
```

```python
intarray = np.arange(6)
print(intarray)
```

```python
intarray.dtype
```

```python
floatarray = np.random.uniform(0,5,5)
print(floatarray)
```

```python
floatarray.dtype
```

```python
concatarray = np.concatenate((intarray,floatarray))
```

```python
print(concatarray)
```

```python
concatarray.dtype
```

```python
downcast = concatarray.astype(int)
```

```python
print(downcast)
```

```python
downcast.dtype
```
