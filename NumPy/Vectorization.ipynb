# Vectorization

```python
import numpy as np
```

```python
a = np.arange(0,5)
```

```python
b = np.arange(5,10)
```

```python
res = []
for i in range(len(a)):
    res.append(a[i]+b[i])
%timeit np.array(res)
```

##### Now we use vectorization

```python
%timeit a + b
```
