## Dictionaries
A dictionary in Python is an unordered, mutable collection of data stored in key–value pairs.
Dictionaries are created using curly braces { } with each item written as key: value.

## Key Points

1. Key–value structure → every value is accessed using a key

2. Mutable → you can add, update, or remove items

Keys must be unique

3. Keys must be immutable types (string, int, tuple, etc.)

4. Values can be any type

```python
# How to Create Dictionarie
Grocery_Bag = {}
print(type(Grocery_Bag))
```

```python
Customers = {
    'cust_id' : 101,
    'first_name' : 'Prafull',
    'last_name' : 'Wahatule',
    'address' : 'Ashti, Beed Maharashtra'
}
print(Customers)
```

```python
print(Customers['cust_id'])
```

```python
print("%d %s %s %s" % (
    Customers['cust_id'],
    Customers['first_name'],
    Customers['last_name'],
    Customers['address']
))
```
