# Set
A set in Python is an unordered, mutable collection of unique elements.
You create a set using curly braces { } or the set() function.

## Key Points

1. Unordered → items do not have a fixed position

2. Unique items only → duplicates are automatically removed

3. Mutable → you can add or remove items

4. Cannot contain mutable items like lists

```python
fruits = {'Apple', 'Mango','Banana'}
```

```python
print(type(fruits))
```

```python
fruits1 = set(('Apple', 'Mango','Banana'))
print(type(fruits))
```

```python
fruits.add('Cheri')
```

```python
print(fruits)
```

```python
fruits.remove('Cheri')
print(fruits)
```

```python
len(fruits)
```

```python
'Mango' in fruits
```

```python
'hahaha' in fruits
```

```python
fruits.add(816)
print(fruits)
```

```python
fruits1.clear()
print(fruits1)
```
