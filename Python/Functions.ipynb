# Functions

A function is a block of reusable code that performs a specific task.
You define a function once and can use (call) it many times in your program.

Functions help make code:

Organized

Reusable

Easy to understand

Easy to maintain


## How to define a function

Use the def keyword.

Example:
def greet():
    print("Hello, Python!")

Calling the function:
greet()

Function with parameters
def add(a, b):
    return a + b

Calling:
print(add(5, 3))   # Output: 8

# User Defined Functions

A User-Defined Function is a function that you create yourself in Python to perform a specific task.
You design it using the def keyword and call it whenever needed.

It helps make your code:

Reusable

Cleaner

Easy to manage


*** A user-defined function is your own custom function created to do a job in your program. ***

```python
# Function Name
def greet():
    print("Hello, welcome!")
```

```python
#Call Function
greet()
```

```python
# Function Name
def add(a, b):
    return a + b
```

```python
#Call Function
add(10,20)
```

```python
fruits = ['Apple','Banana','Mango']
```

```python
def printfriuts(fruitsinfo):
    for inx, fruits in enumerate(fruitsinfo):
        print('Item', inx + 1, 'we are having:', fruits)
```

```python
printfriuts(fruits)
```

```python
numlist = [1,2,3,4,5,6,7,8,9]
```

```python
def squarnum(numberlist):
    squarlist = []

    for number in numberlist:
        x = number**2
        squarlist.append(x)
    return squarlist
```

```python
squarnum(numlist)
```
