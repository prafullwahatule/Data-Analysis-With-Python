# Collection Type - List
A list in Python is an ordered, mutable collection of items.
Lists are created using square brackets [ ], and items are separated by commas.

## Key Points

1. Ordered → items maintain their sequence

2. Mutable → you can add, change, or remove items

3. Can store different data types in the same list

4. Allows duplicate values

# List Comprehensions

```python
somenum = [1,2,3,4,5]
square =[]

for x in somenum:
    square.append(x**2)
    print(x)
print(square)
```

```python
square2 = [x**2 for x in somenum]
print(square2)
```

```python
even_square2 = [x**2 for x in somenum if x%2 == 0]
print(even_square2)
```
