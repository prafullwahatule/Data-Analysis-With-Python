# Classes &n object

# Class – Definition

A class is a blueprint (design) for creating objects.
It defines properties (variables) and behaviors (functions/methods) of the objects.

- In simple words:
A class is like a template.

# Object – Definition

An object is an instance of a class.
It is created from the class and has its own data.

- In simple words:
An object is a copy made from the template.

-- Example
Creating a class
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

--  Creating an object
stud1 = Student("Prafull", 22)

Accessing object data
print(stud1.name)   # Output: Prafull
print(stud1.age)    # Output: 22

= Easy Understanding Example
Class = Car design
Object = Actual car made from that design

```python
class Employee:
    __id = 0
    __name = ""
    __gender = ""
    __age = ""

    def __init__(self,empid,name,gender,age):
        self.__id = empid
        self.__name = name
        self.__gender = gender
        self.__age = age

    def showemployee(self):
        print("Employee Id - ", self.__id)
        print("Employee name - ", self.__name)
        print("Employee gender - ", self.__gender)
        print("Employee age - ", self.__age)
```

```python
emp1 = Employee("101", "Prafull", "Male", "23")
```

```python
print(type(emp1))
```

```python
emp1.showemployee()
```
