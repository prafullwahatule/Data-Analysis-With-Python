## Basic Data Types

1. int (Integer)

An integer is a whole number without decimals.
It can be positive, negative, or zero.

Example: 5, -10, 0

2. float (Floating-point number)

A float is a number that contains a decimal point.

Example: 3.14, 99.5

3. str (String)

A string is a sequence of characters (text) written inside quotes.

Example: "Hello", 'Python'

4. bool (Boolean)

A boolean represents either True or False.

Example: True, False

5. NoneType

None represents the absence of a value or no data.

Example: value = None
