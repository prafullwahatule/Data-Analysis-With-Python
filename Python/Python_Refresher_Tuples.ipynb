# Tuple
A tuple is an ordered, immutable (cannot be changed) collection of items in Python.
You create a tuple by placing items inside parentheses ( ), separated by commas.

## Key Points

1. Ordered → items have fixed positions

2. Immutable → you cannot add, remove, or modify items after creation

3. Can store different data types (int, string, float, etc.)

4. Faster than lists

```python
tuple1 = ()
print(type(tuple1))
```

```python
tuple1 = ('Apple', 'Mango', 'Banana')
print(tuple1)
```

```python
tuple1[0],tuple1[1],tuple1[2]
```

```python
tuple2 = 'Apple', 'Mango', 'Banana'
print(type(tuple2))
print(tuple2)
```

```python
tuple2[0] = 'Orange'
```

```python
tuple3 = (['Apple', 'Mango', 'Banana'],['Applee', 'Mangoo', 'Bananae'])
print(type(tuple3))
print(tuple3)
```

```python
tuple3[0][0] = 'Orange'
print(tuple3)
```

```python
for fruits in tuple2:
    print(fruits.upper())
```
