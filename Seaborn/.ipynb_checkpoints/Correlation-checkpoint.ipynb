## Scatter Plot

### Step 1: Initialize Global Setting

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\midwest.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\midwest.csv")
df
```

### Step 3: Prepare Data

```python
# Unique Category

categories = np.unique(df['category'])
clen = len(categories)
clen

colours = [plt.cm.tab10(i / float(clen - 1)) for i in range(clen)]
colours
```

### Step 4: Draw Plot

```python
# Create a new figure with custom size, resolution, and colors
plt.figure(figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')

# Loop through each unique category with its index
for i, category in enumerate(categories):
    
    # Plot scatter points for the current category only
    plt.scatter(
        "area",                    # X-axis column name
        "poptotal",                # Y-axis column name
        data=df.loc[df.category == category, :],  # Filter data by category
        s=20,                      # Size of each scatter point
        c=[colours[i]],            # Assign a unique color per category
        label=category             # Label for legend
    )

# Set X-axis label
plt.xlabel("Area")

# Set Y-axis label
plt.ylabel("Total Population")

# Set plot title
plt.title("Midwest Area vs Population by Category")

# Display legend for categories
plt.legend()

# Render the plot
plt.show()
```
