# Change Plot

Change Plot is a type of data visualization used to show how a value changes over time or between different states.
It focuses on the difference, increase, decrease, or trend in data rather than just the actual values.

### In simple words:

- A Change Plot shows how data moves from one point to another.

- Key Characteristics

- Highlights variation or trend

- Shows increase, decrease, or stability

- Compares before vs after values

- Often uses lines, arrows, or slope to indicate change

## 1. Change Plot : Time Series Plot

A Time Series Plot is a type of Change Plot that displays how a variable changes over time by plotting time on the X-axis and the measured value on the Y-axis.

#### Key Points

- X-axis represents time (days, months, years, etc.)

- Y-axis represents the value of the variable

- Shows trend, pattern, seasonality, and fluctuations

- Usually drawn using a line chart

#### Purpose

- To observe increase or decrease over time

- To identify trends and seasonal patterns

- To analyze past behavior and predict future values

## 2. Change Plot : Time Series Decomposition Plot

A Time Series Decomposition Plot is a type of Change Plot that breaks a time series into its individual components to clearly understand how the data changes over time.

#### Components of Time Series Decomposition

1ï¸. Trend (T)

- Shows the long-term movement of the data

- Indicates overall increase or decrease over time

ï¸2. Seasonality (S)

- Repeating periodic patterns

- Occurs at fixed intervals (monthly, quarterly, yearly)

ï¸3. Residual / Noise (R)

- Random fluctuations not explained by trend or seasonality

- Types of Decomposition

- Additive Model

Y=T+S+R

- Used when seasonal variation is constant over time.

- Multiplicative Model

Y=TÃ—SÃ—R

- Used when seasonal variation changes with level.

### Purpose

- To understand data behavior

- To identify hidden patterns

- To analyze trend and seasonality separately

- To improve forecasting accuracy

### Step 1: Initialize Global Settings

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
import squarify
import random

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\AirPassengers.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\AirPassengers.csv")
df
```

### Step 3: Prepare Data

```python
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
```

### Step 4: Draw Plot

```python
# Decompose the time series (monthly data)
result = seasonal_decompose(
    df["value"],
    model="additive",
    period=12
)

# Plot decomposition
result.plot()
plt.suptitle("Time Series Decomposition Plot", fontsize=18)
plt.show()
```

```python
# Plot decomposition
result.plot()
plt.suptitle("Time Series Decomposition Plot", fontsize=18)
plt.show()
```
