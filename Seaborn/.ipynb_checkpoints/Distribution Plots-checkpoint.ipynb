# Distribution Plots

Distribution plots show how data values are spread across a range. They help you understand the shape, center, spread, and outliers of a dataset.

## Distribution Plots : Histogram for Continuous Variable

A Histogram is one of the most common plots used to understand the distribution of a continuous variable. It shows how data values are spread across different ranges (bins).

#### What a Histogram Shows

- X-axis: Continuous variable (e.g., age, income, mpg)

- Y-axis: Frequency (count of observations in each bin)

- Bars: Represent how many values fall into each interval (bin)

#### Why Use a Histogram?

- A histogram helps you understand:

- Data distribution (normal, skewed, uniform)

- Central tendency (where most values lie)

- Spread / variability

- Skewness (left or right skew)

- Outliers

- Modality (uni-modal, bi-modal, multi-modal)

### Step 1: Setting Globale 

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
x_var = "displ"
groupby_var = "class"

df_agg = df.loc[:,[x_var,groupby_var]].groupby(groupby_var)
vals = [df[x_var].values.tolist() for i, df in df_agg]
```

### Step 4: Draw Plot

```python
# Colors
colors = [plt.cm.Spectral(i / float(len(vals) - 1))
          for i in range(len(vals))]

# Plot
plt.figure(figsize=(16, 10), dpi=80)

n, bins, patches = plt.hist(
    vals,
    bins=30,
    stacked=True,
    density=False,
    color=colors
)

# Legend
plt.legend(
    {group: col for group, col in zip(df[groupby_var].unique(), colors)},
    title=groupby_var
)

# Title and labels
plt.title(
    f"Stacked Histogram of {x_var} colored by {groupby_var}",
    fontsize=22
)
plt.xlabel(x_var)
plt.ylabel("Frequency")

# Y-axis limit
plt.ylim(0, 25)

# X-ticks
plt.xticks(
    ticks=bins[::3],
    labels=[round(b, 1) for b in bins[::3]]
)

plt.show()

```

### Step 5: decorate Plots With Right Labels

```python
# Legend
plt.legend(
    {group: col for group, col in zip(df[groupby_var].unique(), colors)},
    title=groupby_var
)

# Title and labels
plt.title(
    f"Stacked Histogram of {x_var} colored by {groupby_var}",
    fontsize=22
)
plt.xlabel(x_var)
plt.ylabel("Frequency")

# Y-axis limit
plt.ylim(0, 25)

# X-ticks
plt.xticks(
    ticks=bins[::3],
    labels=[round(b, 1) for b in bins[::3]]
)

plt.show()
```

## Distribution Plot : Density Plot

A Density Plot is a smooth, continuous curve that represents the probability distribution of a continuous variable. It is a refined version of a histogram that removes the dependence on bin width.

### Why use a Density Plot?

- Shows the shape of data clearly

- Easier to compare multiple distributions

- Highlights skewness, peaks (modes), and spread

- Smoother and more informative than histograms

### Key Characteristics

- Area under the curve = 1

- X-axis â†’ variable values

- Y-axis â†’ density (probability)

- Works only for continuous variables

## Distribution Plot : Box Plot

A Box Plot is a statistical visualization that summarizes the distribution of a dataset using five key numbers and clearly highlights outliers.

### What a Box Plot Shows (Five-Number Summary)

- Minimum (excluding outliers)

- Q1 (25th percentile)

- Median (Q2 / 50th percentile)

- Q3 (75th percentile)

- Maximum (excluding outliers)

ðŸ“¦ Box â†’ IQR = Q3 âˆ’ Q1
ðŸ“ Whiskers â†’ extend up to 1.5 Ã— IQR
ðŸ”´ Dots â†’ Outliers

### Why Use a Box Plot?

1. Understand spread & central tendency

2. Detect outliers

3. Compare multiple groups

### Step 1: Setting Globale 

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 4: Draw Plot

```python
def add_n_obs(df, group_col, y):
    # Median of each group
    medians_dict = {
        grp[0]: grp[1][y].median()
        for grp in df.groupby(group_col)
    }

    # X-axis tick labels
    xticklabels = [x.get_text() for x in plt.gca().get_xticklabels()]

    # Number of observations per group
    n_obs = df.groupby(group_col)[y].size().values

    # Add text annotations
    for x, (xticklabel, n_ob) in enumerate(zip(xticklabels, n_obs)):
        plt.text(
            x,
            medians_dict[xticklabel] * 1.01,
            f"#obs: {n_ob}",
            horizontalalignment="center",
            fontsize=12,
            color="black"
        )

plt.figure(figsize=(16, 10))

sns.boxplot(
    x="class",
    y="hwy",
    data=df
)

add_n_obs(df, group_col="class", y="hwy")

plt.title("hwy Mileage by Vehicle Class", fontsize=18)
plt.xlabel("Vehicle Class")
plt.ylabel("City Mileage")
plt.show()
```
