# Composition Plots

Composition plots are used to show how a whole is divided into parts. They answer questions like:

- What proportion does each category contribute to the total?

- How does the composition change over time or across groups?

## Composition Plots : Pie Chart

A Pie Chart is a composition plot used to show how a whole (100%) is divided into parts.

### What it Shows

1. Proportion / percentage of each category

2. Relationship of parts to the whole

### When to Use

✔ Few categories (ideally ≤ 5)
✔ To show percentage contribution
❌ Not good for comparing very similar values

### Step 1: Initialize Global Setting

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
# Grouping
df_processed = (
    df.groupby("class")
      .size()
      .reset_index(name="count")
)
```

### Step 4: Draw Plot

```python
# Figure
fig, ax = plt.subplots(
    figsize=(12, 7),
    subplot_kw=dict(aspect="equal"),
    dpi=80
)

# Data
data = df_processed["count"]
categories = df_processed["class"]

explode = [0, 0, 0, 0, 0, 0.1, 0]

# Pie chart
ax.pie(
    data,
    labels=categories,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True
)

ax.set_title("Composition Plot : Vehicle Class Distribution", fontsize=18)
plt.show()
```

### Step 5: Decorate plots with Right Labels

```python
ax.set_title("Composition Plot : Vehicle Class Distribution", fontsize=18)
plt.show()
```

## Composite Plot : Tree Map

A Treemap is a composition plot used to show part-to-whole relationships, especially when there are many categories or hierarchical data.

### What a Treemap Shows

- Each rectangle = a category

- Area of rectangle ∝ value

- Efficient replacement for pie charts when categories are many

### When to Use Treemap

✔ Many categories
✔ Hierarchical composition
✔ Dashboard visualizations
❌ Precise value comparison

### Step 1: Initialize Global Setting

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import squarify

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
# Prepare data
df_processed = (
    df.groupby("class")
      .size()
      .reset_index(name="count")
)

sizes = df_processed["count"]
labels = df_processed["class"] + "\n" + df_processed["count"].astype(str)
```

### Step 4: Draw Plot

```python
# Plot
plt.figure(figsize=(12, 7), dpi=80)
squarify.plot(
    sizes=sizes,
    label=labels,
    alpha=0.8
)

plt.title("Composition Plot : Vehicle Class Treemap", fontsize=18)
plt.axis("off")
plt.show()
```

### Step 5: Decorate plots with Right Labels

```python
plt.title("Composition Plot : Vehicle Class Treemap", fontsize=18)
plt.axis("off")
plt.show()
```

## Composite Plot : Bar Chart
A Composition Bar Chart shows how a total is divided into parts using bar lengths instead of angles (pie chart).
It is often clearer and more accurate than a pie chart.

### What It Shows

- Part-to-whole relationship

- Contribution of each category to the total

### When to Use

✔ Many categories
✔ Easy comparison between parts
✔ When pie charts become cluttered

### Step 1: Initialize Global Setting

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import squarify

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
df_processed = (
    df.groupby("class")
      .size()
      .reset_index(name="count")
)
```

### Step 4: Draw Plot

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi=80)

ax.bar(
    ["All Vehicles"],
    [100]
)

bottom = 0
for _, row in df_processed.iterrows():
    ax.bar(
        ["All Vehicles"],
        row["percentage"],
        bottom=bottom,
        label=row["class"]
    )
    bottom += row["percentage"]

ax.set_title("100% Composition Bar Chart")
ax.set_ylabel("Percentage")
ax.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

### Step 5: Decorate plots with Right Labels

```python
ax.set_title("100% Composition Bar Chart")
ax.set_ylabel("Percentage")
ax.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
