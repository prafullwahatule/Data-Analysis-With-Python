# Deviation Plot

A Deviation Plot shows how each data point deviates from a reference value (usually the mean or median). It’s very useful to quickly see above-average vs below-average values.

### What a Deviation Plot Shows

- Center line → Mean (or median)

- Points above the line → Positive deviation

- Points below the line → Negative deviation

- Distance from the line → Magnitude of deviation

## Deviation Plot : Diverging Bars

A Diverging Bar Deviation Plot shows how values deviate from a reference point (usually the mean) using bars that diverge left and right (or up/down).
It clearly separates positive vs negative deviation.

### Concept

- Center line (0) → Mean (or baseline)

- Bars to the right / up → Above average

- Bars to the left / down → Below average

- Length of bar → Magnitude of deviation

### Step 1: Initialize Global Settings

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mtcars.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mtcars.csv")
df
```

### Step 3: Prepare Data

```python
x = df.loc[:,["mpg"]]
df["mpgz"] - (x-x.mean()/x.std())

df["colors"] = ["red" if x < 0 else "green" for x in df["mpgz"]]
df.sort_values("mpgz", inplace = True)
df.reset_index(inplace = True)
```

### Step 4: Draw Plot

```python
plt.figure(figsize=(16, 10), dpi=80)

plt.hlines(
    y=df.index,
    xmin=0,
    xmax=df["mpgz"],
    color=df["colors"],
    alpha=0.4,
    linewidth=5
)


# Dots at the end
plt.scatter(
    df["mpgz"],
    df.index,
    color=df["colors"],
    s=80,
    zorder=3
)

plt.axvline(0, color="black", linewidth=1)

plt.gca().set(
    ylabel="Model",
    xlabel="Mileage Deviation (Z-score)"
)

plt.yticks(df.index, df["cars"], fontsize=12)

plt.title("Deviation Plot – Diverging Dot Plot of Car Mileage")

plt.grid(linestyle="--", alpha=0.5)

plt.show()
```

## Deviation Plot : Diverging Dot Plot

A Diverging Dot Plot visualizes how each value deviates from a reference point (usually the mean or median) using dots instead of bars.
It’s cleaner than bars and great when you have many observations.

### Key Idea

- Zero line (0) → Reference (mean / median)

- Dots right of 0 → Positive deviation

- Dots left of 0 → Negative deviation

- Distance from 0 → Strength of deviation

### Step 1: Initialize Global Settings

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mtcars.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mtcars.csv")
df
```

### Step 3: Prepare Data

```python
df["mpgz"] = (df["mpg"] - df["mpg"].mean()) / df["mpg"].std()

df["colors"] = np.where(df["mpgz"] < 0, "red", "green")

df.sort_values("mpgz", inplace=True)
df.reset_index(drop=True, inplace=True)
```

### Step 4: Draw Plot

```python
# Dots at the end (recommended)
plt.scatter(
    df["mpgz"],
    df.index,
    color=df["colors"],
    s=80,
    zorder=3
)

plt.axvline(0, color="black", linewidth=1)

plt.gca().set(
    ylabel="Model",
    xlabel="Mileage Deviation (Z-score)"
)

plt.yticks(df.index, df["cars"], fontsize=12)

plt.title("Deviation Plot – Diverging Dot Plot of Car Mileage")

plt.grid(linestyle="--", alpha=0.5)

plt.show()
```
