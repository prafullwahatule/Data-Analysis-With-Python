## Scatter Plot

A scatter plot is used to show the relationship between two numerical variables.
Each point represents one observation in the dataset.

### Step 1: Initialize Global Setting

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\midwest.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\midwest.csv")
df
```

### Step 3: Prepare Data

```python
# Unique Category

categories = np.unique(df['category'])
clen = len(categories)
clen

colours = [plt.cm.tab10(i / float(clen - 1)) for i in range(clen)]
colours
```

### Step 4: Draw Plot

```python
plt.figure(figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')

for i, category in enumerate(categories):
    plt.scatter(
        "area",
        "poptotal",
        data=df.loc[df.category == category, :],
        s=20,
        c=[colours[i]],
        label=category
    )

plt.xlabel("Area")
plt.ylabel("Total Population")
plt.title("Midwest Area vs Population by Category")
plt.legend()
plt.show()
```

## Scatter Plot With best Fit Line

A best-fit line (regression line) shows the overall trend between two numerical variables.

### Step 1: Initialize Global Setting

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
## "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"


df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
dfs = df.loc[df.cyl.isin([4,8])]
dfs
```

### Step 4: Draw Plot

```python
# Set seaborn background style to white for a clean look
sns.set_style("white")

# lmplot creates a scatter plot with a regression (best-fit) line
# It is a figure-level function based on FacetGrid
plotobj = sns.lmplot(
    x="displ",            # X-axis: engine displacement
    y="hwy",              # Y-axis: highway mileage
    hue="cyl",            # Group data by number of cylinders (color + line)
    data=dfs,             # Dataset used for plotting
    height=7,             # Height of the figure (in inches)
    aspect=1.6,           # Width = height Ã— aspect
    robust=True,          # Uses robust regression (less sensitive to outliers)
    palette="tab10",      # Color palette for different cylinder groups
    scatter_kws=dict(     # Styling options for scatter points
        s=60,             # Size of each point
        linewidths=0.7,   # Border thickness of points
        edgecolors="black"  # Black outline for better visibility
    )
)

# Set X and Y axis limits
plotobj.set(xlim=(0.5, 7.5), ylim=(0, 50))

# Add a title to the entire figure (correct way for lmplot)
plotobj.fig.suptitle(
    "Scatter Plot with Line of Best Fit Grouped by Number of Cylinders",
    fontsize=large
)
```

## Correlation : Count Plot

A count plot is NOT used to measure correlation in the statistical sense.
Instead, it is used to compare frequencies between categories.

### Step 1: Initialize Global Setting

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
## "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"


df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
df_counts = df.groupby(["hwy","cty"]).size().reset_index(name = "counts")
df_counts
```

### Step 4: Draw Plot

```python
fig, ax = plt.subplots(figsize=(16, 10), dpi=80)

ax.scatter(
    df_counts['cty'],
    df_counts['hwy'],
    s=df_counts['counts'] * 20,  # size represents frequency
    alpha=0.6
)

ax.set_xlabel("City Mileage")
ax.set_ylabel("Highway Mileage")
ax.set_title("Bubble Plot: Frequency of City vs Highway Mileage")
plt.show()
```

## Correlation Distribution : Marginal Boxplot

Marginal boxplots are used to visualize both the correlation between two numerical variables and the distribution of each variable along its axis.

- Scatter: correlation/trend

- Marginal box: spread, median, outliers

### Step 1: Initialize Global Setting



```python
  # Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
## "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"


df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
# Create the main figure
fig = plt.figure(figsize=(16, 10), dpi=80)

# Create a GridSpec layout (4 rows Ã— 4 columns)
grid = plt.GridSpec(4, 4, hspace=0.5, wspace=0.2)

# Main scatter plot (top-left area)
ax_main = fig.add_subplot(grid[:-1, :-1])

# Right marginal plot (Y distribution)
ax_right = fig.add_subplot(
    grid[:-1, -1],
    xticklabels=[],
    yticklabels=[]
)

# Bottom marginal plot (X distribution)
ax_bottom = fig.add_subplot(
    grid[-1, 0:-1],
    xticklabels=[],
    yticklabels=[]
)

# Main scatter plot (correlation)
ax_main.scatter(
    df['displ'],
    df['hwy'],
    c=df['manufacturer'].astype('category').cat.codes,  # color by manufacturer
    cmap='Set1',
    alpha=0.9,
    edgecolors='black',
    linewidth=0.5
)

# Bottom marginal boxplot (distribution of displ)
sns.boxplot(
    x=df['displ'],
    ax=ax_bottom,
    orient='h'
)

# Right marginal boxplot (distribution of hwy)
sns.boxplot(
    y=df['hwy'],
    ax=ax_right,
    orient='v'
)


# ===============================
# Titles & labels
# ===============================
ax_main.set_title("Scatter Plot with Marginal Boxplots")
ax_main.set_xlabel("Engine Displacement (displ)")
ax_main.set_ylabel("Highway Mileage (hwy)")

plt.show()
```

### Step 4: Draw Plot

```python
# Main scatter plot (correlation)
ax_main.scatter(
    df['displ'],
    df['hwy'],
    c=df['manufacturer'].astype('category').cat.codes,  # color by manufacturer
    cmap='Set1',
    alpha=0.9,
    edgecolors='black',
    linewidth=0.5
)

# Bottom marginal boxplot (distribution of displ)
sns.boxplot(
    x=df['displ'],
    ax=ax_bottom,
    orient='h'
)

# Right marginal boxplot (distribution of hwy)
sns.boxplot(
    y=df['hwy'],
    ax=ax_right,
    orient='v'
)

ax_main.set
```

## Correlation : Correlogram Heatmap

### Step 1: Initialize Global Setting

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import DatasetÂ¶

```python
## "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mtcars.csv"


df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mtcars.csv")
df
```

### Step 3: Prepare Data

```python
num_df = df.select_dtypes(include='number')
num_df.dtypes


corr = num_df.corr()
corr
```

### Step 4: Draw Plot

```python
# mask upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(16, 10))
sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="RdBu_r",
    vmin=-1,
    vmax=1,
    center=0,
    square=True,
    linewidths=0.5
)

plt.title("Correlation Correlogram â€“ mtcars Dataset")
plt.show()
```
