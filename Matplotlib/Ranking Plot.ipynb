# Ranking Plot

A Ranking Plot is used to rank items based on a numeric value and visualize their order (highest to lowest or vice-versa).
Common use cases:

- Ranking products by sales

- Ranking students by marks

- Ranking cars by mileage (mpg)

- Top / Bottom N analysis

## Ordered Bar Plot

An Ordered Bar Plot is a bar chart where bars are sorted by their values (ascending or descending).
It helps you quickly see ranking, patterns, and extremes.

### Why use an Ordered Bar Plot?

✔ Highlights highest & lowest values
✔ Easier comparison than unordered bars
✔ Common in EDA & dashboards

### Step 1: Initialize Global Settings

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
processeddf = df[["cty","manufacturer"]].groupby("manufacturer").apply(lambda x:x.mean())
processeddf.sort_values("cty",inplace = True)
processeddf.reset_index(inplace = True)
```

### Step 4: Draw Plot

```python
fig, ax = plt.subplots(
    figsize=(12, 6),
    facecolor="white",
    dpi=80
)

ax.vlines(
    x=processeddf.index,
    ymin=0,
    ymax=processeddf["cty"],
    color="firebrick",
    alpha=0.7,
    linewidth=20
)

for i, cty in enumerate(processeddf["cty"]):
    ax.text(
        i,
        cty + 0.5,
        round(cty, 1),
        horizontalalignment="center"
    )

ax.set_title("Bar Chart for City Mileage", fontsize=22)
ax.set(
    ylabel="Miles per Gallon",
    ylim=(0, 30)
)

# Convert manufacturer names to UPPERCASE
labels = processeddf["manufacturer"].str.upper()
plt.xticks(processeddf.index, labels, rotation=90)

plt.show()
```

### Step 5: Decprate Plots with Right Labels

## Ranking Plot : Dot Plot

A Ranking Dot Plot (also called a Cleveland Dot Plot) is used to rank categories by a numeric value using dots instead of bars.
It is cleaner, more precise, and less cluttered than bar charts.

#### When to Use a Ranking Dot Plot

✔ Many categories
✔ Focus on rank + exact value
✔ Comparing items clearly
✔ Better alternative to bar charts in EDA

### Step 1: Initialize Global Settings

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# setting up the params
large = 22
med = 12
small = 12

params = {
    'axes.titlesize': large,
    'legend.fontsize': med,
    'figure.figsize': (16, 10),
    'axes.labelsize': med,
    'xtick.labelsize': med,
    'ytick.labelsize': med,
    'figure.titlesize': large
}

# override default parameters of plot
plt.rcParams.update(params)

# set style (updated seaborn style)
plt.style.use('seaborn-v0_8-whitegrid')
```

### Step 2: Import Dataset

```python
# "C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv"

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Python_DA\Data\mpg.csv")
df
```

### Step 3: Prepare Data

```python
processeddf = df[["cty","manufacturer"]].groupby("manufacturer").apply(lambda x:x.mean())
processeddf.sort_values("cty",inplace = True)
processeddf.reset_index(inplace = True)
```

### Step 4: Draw Plot

```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6), dpi=80)

# -----------------------------
# Draw horizontal reference lines (sticks)
# -----------------------------
# y  -> category positions (index)
# xmin -> starting point of line (0)
# xmax -> value of city mileage
ax.hlines(
    y=processeddf.index,
    xmin=0,
    xmax=processeddf["cty"],
    color="grey",
    alpha=0.4
)

# -----------------------------
# Draw dots at the end of each line
# -----------------------------
# x -> city mileage values
# y -> category positions
ax.plot(
    processeddf["cty"],
    processeddf.index,
    "o",
    color="firebrick"
)

# -----------------------------
# Set y-axis labels (manufacturer names)
# -----------------------------
ax.set_yticks(processeddf.index)
ax.set_yticklabels(processeddf["manufacturer"].str.upper())

# -----------------------------
# Axis labels and title
# -----------------------------
ax.set_xlabel("City Mileage (MPG)")
ax.set_ylabel("Cars")
ax.set_title("Ranking Plot : Dot Plot (City Mileage)", fontsize=20)

# -----------------------------
# Improve readability
# -----------------------------
ax.grid(axis="x", linestyle="--", alpha=0.4)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Display plot
plt.show()
```
