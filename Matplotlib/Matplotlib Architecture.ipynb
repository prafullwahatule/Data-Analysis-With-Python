# Matplotlib Architecture


Matplotlib is a Python library used for creating visualizations (charts, graphs).
Its architecture is built in layers so you can control your plots with flexibility.

#### Matplotlib uses the concept of:
- Figure
- Axes
- Axis
- Artist

### Main Components of Matplotlib Architecture
1️. Figure

- The entire canvas or window where your plot appears.

- It can contain one or multiple plots.

Ex : fig = plt.figure()

2️. Axes

- A single plot area inside the figure.

- Contains:

  - Title
  - X-axis
  - Y-axis
  - Actual graph (lines, bars, etc.)
EX : ax = fig.add_subplot(1, 1, 1)

3️. Axis

- These are x-axis and y-axis of the plot.

- They include:

    - Ticks (values like 0, 10, 20…)

    - Tick labels

    - Gridlines

EX : 
ax.set_xlabel("Time")
ax.set_ylabel("Sales")

4️. Artist

- Everything visible in a plot is an Artist.

- Examples of artists:

    - Line

    - Bar

    - Text

    - Title

    - Legend

    - Grids

Matplotlib builds a plot by drawing artists on the axes.

### Matplotlib Architecture Diagram (Text Version)
Figure
 └── Axes (Plot Area)
       ├── Axis (X-axis)
       ├── Axis (Y-axis)
       └── Artists (Lines, Bars, Titles, Labels, etc.)

### Summary
Component	Meaning
Figure	Entire drawing area
Axes	Individual graph area (where lines appear)
Axis	X and Y axis of the plot
Artist	Everything drawn on the plot


```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot([1,2,3], [10,20,30])
ax.set_title("Simple Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```
