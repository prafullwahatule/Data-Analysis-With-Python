# Parameters of Plot

When we create a plot using Matplotlib, we can customize many things.
These customizations are called plot parameters.

The most commonly used parameters are:

### 1. Title

Adds a title to the plot.

plt.title("Sales Trend")

### 2. Labels (X-label & Y-label)

Describe what each axis represents.

plt.xlabel("Months")
plt.ylabel("Revenue")

### 3. Color

Used to set line, bar, scatter color.

plt.plot(x, y, color='red')

### 4. Line Style

Changes the style of the line.

Style	Code
Solid	'-'
Dotted	':'
Dashed	'--'
Dash-dot	'-.'
plt.plot(x, y, linestyle='--')

### 5. Line Width

Thickness of the line.

plt.plot(x, y, linewidth=3)

### 6. Marker

Adds markers at each data point.

Marker	Code
Circle	'o'
Square	's'
Triangle	'^'
Star	'*'
plt.plot(x, y, marker='o')

### 7. Marker Size & Color

Control marker appearance.

plt.plot(x, y, marker='o', markersize=8, markerfacecolor='yellow')

### 8. Grid

Add reference lines.

plt.grid(True)

### 9. Legend

Shows labels for multiple lines.

plt.legend(["Sales", "Profit"])

### 10. Figure Size

Sets the width and height of the plot.

plt.figure(figsize=(8, 5))

### 11. Axis Limits

Control the visible range on axes.

plt.xlim(0, 100)
plt.ylim(0, 500)

### Summary
Parameter	        - Purpose
1. title	        - Add title
2. xlabel, ylabel	- Label axes
3. color	        - Line/bar color
4. linestyle	    - Style of line
5. linewidth	    - Thickness of line
6. marker	        - Mark data points
7. markersize	    - Size of marker
8. grid	            - Background grid
9. legend	        - Label series
10. figsize	        - Change plot size
11. xlim, ylim	    - Axis range
