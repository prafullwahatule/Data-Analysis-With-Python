# Matplotlib

### Learn by Doing Exercises

1. How to draw Curve using matplotlib
2. How to draw multiple curve on the same plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## How to draw Curve using matplotlib

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
## How to draw multiple curve on the same plot

a = np.linspace(0,2*np.pi,100)
b = np.sin(a)
c = np.cos(a)
plt.plot(a,b)
plt.plot(a,c)
```
