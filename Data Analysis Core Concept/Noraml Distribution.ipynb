# Noraml Distribution

A Normal Distribution (also called Gaussian Distribution) is a type of continuous probability distribution where:

Most of the data points are clustered around the mean

Fewer values appear as you move away from the mean

The graph forms a smooth, symmetric bell-shaped curve

### Key Characteristics
1️. Bell-Shaped Curve

The curve is highest at the mean and decreases on both sides.

2️. Symmetry

The left side and right side of the curve are mirror images.

3️. Mean = Median = Mode

All three are equal and located at the center.

4️. Most Data Lies Near the Mean

#### About:

68% of data lies within 1 standard deviation (σ)

95% within 2σ

99.7% within 3σ

This is known as the 68-95-99.7 Rule.

5️. Defined by Mean and Standard Deviation

Mean (μ) → center of the curve

Standard Deviation (σ) → spread of the curve

#### Example

Heights of people, exam scores, and measurement errors often follow a normal distribution.

### Summary (Short Notes)

A normal distribution is a symmetric, bell-shaped curve where most values cluster around the mean, and data spreads based on standard deviation.
