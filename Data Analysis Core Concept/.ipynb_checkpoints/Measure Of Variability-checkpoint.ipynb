# Measure Of Variability

Measures of Variability tell us how spread out or how scattered the data values are in a dataset.

Central tendency (mean, median, mode) tells the center,
while variability tells the spread.

### Why Variability Is Important?

Because two datasets can have the same mean but very different spreads.

Example:

Dataset A :	10, 10, 10, 10	
Dataset B : 1, 10, 20, 29

Both have mean = 10,
but Dataset B is more spread out â†’ higher variability.
