# Modality Skewness Kurtosis

These three concepts describe the shape of a distribution.

## 1. Modality

Modality tells us how many peaks (modes) a distribution has.

### Types of Modality:
Type	    : Meaning	Shape
Unimodal	: 1 peak	Most common
Bimodal	    : 2 peaks	Two groups in data
Multimodal	: 3 or more peaks	Complex distribution


#### Example:

Sales every morning and evening → Bimodal

## 2. Skewness

Skewness measures the asymmetry (tilt) of a distribution.

### Types of Skewness:
#### ️1. Positive Skew (Right Skewed)

Long tail on right side

Mean > Median

Example: Income distribution

#### 2. Negative Skew (Left Skewed)

Long tail on left side

Mean < Median

Example: Exam scores where most students score high

#### 3️. Zero Skew (Symmetric)

Bell-shaped (Normal distribution)

Mean = Median = Mode

## 3. Kurtosis

Kurtosis measures the tailedness of a distribution
(How heavy or light the tails are).

### Types of Kurtosis:
Type	    : Meaning	: Interpretation  

Mesokurtic  : (K = 3)	: Normal distribution	Moderate tails  

Leptokurtic : (K > 3)	: Heavy tails	More outliers  

Platykurtic : (K < 3)	: Light tails	Fewer outliers  


#### Simple Meaning:

High Kurtosis → data has more extreme values/outliers

Low Kurtosis → data is more flat, fewer outliers
